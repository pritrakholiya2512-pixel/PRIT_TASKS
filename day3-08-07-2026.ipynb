```python
import numpy as np

arr1 = np.array([[1, 2, 3],[4, 5, 6]])
arr2 = np.array([[10, 20, 30],[40, 50, 60]])

print("Array 1:")
print(arr1)

print("\nArray 2:")
print(arr2)

# Indexing
print("\nFirst Element:", arr1[0,0])
print("Last Element:", arr1[1,2])

# Mathematical Operations
print("\nAddition:")
print(arr1 + arr2)

print("\nSubtraction:")
print(arr2 - arr1)

print("\nMultiplication:")
print(arr1 * arr2)

print("\nDivision:")
print(arr2 / arr1)

# 15 NumPy Methods

print("\nShape:", arr1.shape)
print("Dimensions:", arr1.ndim)
print("Size:", arr1.size)
print("Data Type:", arr1.dtype)

print("\nSum:", np.sum(arr1))
print("Mean:", np.mean(arr1))
print("Max:", np.max(arr1))
print("Min:", np.min(arr1))
print("Standard Deviation:", np.std(arr1))

print("\nTranspose:")
print(np.transpose(arr1))

print("\nFlatten:")
print(arr1.flatten())

print("\nSorted Array:")
print(np.sort(arr2))

print("\nSquare Root:")
print(np.sqrt(arr2))

print("\nPower:")
print(np.power(arr1, 2))

print("\nReshape:")
print(arr1.reshape(3,2))
```
