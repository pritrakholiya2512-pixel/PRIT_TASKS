```
#DICTIONARY
student = {
    "name": "Prit",
    "age": 20,
    "city": "Surat"
}

# 1. get()
print(student.get("name"))

# 2. keys()
print(student.keys())

# 3. values()
print(student.values())

# 4. update()
student.update({"course": "Python"})
print(student)

# 5. pop()
student.pop("city")
print(student)
```

```
#LIST
numbers = [10, 20, 30]

# 1. append()
numbers.append(40)
print(numbers)

# 2. insert()
numbers.insert(1, 15)
print(numbers)

# 3. remove()
numbers.remove(20)
print(numbers)

# 4. sort()
numbers.sort()
print(numbers)

# 5. reverse()
numbers.reverse()
print(numbers)
```

```
#TUPLE
fruits = ("Apple", "Banana", "Mango", "Banana")

# 1. count()
print(fruits.count("Banana"))

# 2. index()
print(fruits.index("Mango"))

# 3. len()
print(len(fruits))

# 4. max()
print(max(fruits))

# 5. min()
print(min(fruits))
```

```
#SET
colors = {"Red", "Blue", "Green"}

# 1. add()
colors.add("Yellow")
print(colors)

# 2. remove()
colors.remove("Blue")
print(colors)

# 3. update()
colors.update(["Black", "White"])
print(colors)

# 4. discard()
colors.discard("Pink")
print(colors)

# 5. pop()
colors.pop()
print(colors)
```

```
#IF
age = 20
if age >= 18:
    print("You are eligible to vote.")
```

```
#IF ELSE
num = 5
if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

```
#IF ELIF ELSE
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 70:
    print("Grade B")
else:
    print("Grade C")
```

```
#NESTED IF ELSE
num = 15

if num > 0:
    if num % 2 == 0:
        print("Positive Even")
    else:
        print("Positive Odd")
else:
    print("Negative Number")
```

```
#BREAK
for i in range(1, 11):
    if i == 6:
        break
    print(i)
```

```
#CONTINUE
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

```
#PASS
for i in range(1, 4):
    if i == 2:
        pass
    print(i)
```

```
#INPUT METHOD
name = input("Enter your name: ")
print("Hello,", name)
```

```
#RANGE FUNCTION
for i in range(1, 6):
    print(i)
```

```
#LEN FUNCTION
text = "Python"
print(len(text))
```

```
#TYPE FUNCTION
x = 100
print(type(x))
```

```
#FOR LOOP
for i in range(1, 6):
    print("Number:", i)
```

```
#WHILE LOOP
count = 1
while count <= 5:
    print("Count:", count)
    count += 1
```
